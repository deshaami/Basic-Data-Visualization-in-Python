```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd # Import pandas

x=[1,2,3,4,5]
y=[6,4,7,3,2]

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.figure()

# Create a pandas DataFrame from the lists
data = pd.DataFrame({'X': x, 'Y': y})

# Use the DataFrame with stripplot, specifying the column names for x and y
b = sb.stripplot(x='X', y='Y', data=data)

b.set(xlabel='X-axis',ylabel='Y-axis')
plt.show()
```
